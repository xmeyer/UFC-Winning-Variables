# Exploratory Data Analysis
* Select variables 
 - red fighter
 - winner (boolean: red = 1, blue = 0)
 - stance
 - height cms
 - reach cms
 - weight lbs
 - age

```python
red = pd.DataFrame(f, columns = ['R_fighter', 'Winner', 'R_Stance',
                                 'R_Height_cms', 'R_Reach_cms',
                                 'R_Weight_lbs', 'R_age'])
red.head()
```

```python
print(red.shape, '\n', red.isnull().sum())
```

```python
# Remove missing data
red = red.dropna()
print(red.shape, '\n', red.isnull().sum())
```

```python
# Winner: win = 1, loss = 0, draw = 2 (remove draw?)
red['Winner'] = red['Winner'].replace({'Red': 1,
                                       'Blue': 0,
                                       'Draw': 2}).astype(int)

# Stance: Orthodox = 1, Southpaw = 0, Switch = 2, Open Stance = 3
red['R_Stance'] = red['R_Stance'].replace({'Orthodox': 1,
                                           'Southpaw': 0,
                                           'Switch': 2,
                                           'Open Stance': 3}).astype(int)

# Height and Reach: cms to m
red['R_Height_m'] = red['R_Height_cms'] / 100
red['R_Reach_m'] = red['R_Reach_cms'] / 100

# Weight: lbs to kg
red['R_Weight_kg'] = red['R_Weight_lbs'] * 0.453592

# Remove unwanted columns
red = red.drop(columns = ['R_Height_cms', 'R_Reach_cms', 'R_Weight_lbs'])

# BMI: weight/ height^2
red['R_bmi'] = red['R_Weight_kg'] / (red['R_Height_m'] ** 2)

# Round decimals to hundredths
red = red.round(2)

red.head()
```

```python
red['R_Stance'].value_counts()
```

```python
# Subplots of stance (hist/ bar vs wins), age, height, reach, weight, bmi
red['R_Stance'].plot.hist();
```

```python
# Change style
plt.style.use('seaborn-whitegrid')

# Subplots of stance (hist/ bar vs wins), age, height, reach, weight, bmi
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows = 2,
                                         ncols = 2,
                                         figsize = (10, 10))


# Add data to ax0
scatter = ax0.scatter(x = red['R_age'],
                      y = red['R_bmi'],
                      c = red['Winner'],
                      cmap = 'winter')
# Customize ax0
ax0.set(title = 'Fitness (Age v. BMI)',
        xlabel = 'Age',
        ylabel = 'BMI');
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = 'Winner')
# Add a meanline
ax0.axhline(y = red['R_bmi'].mean(),
            linestyle = '--');


# Add data to ax1
scatter = ax1.scatter(x = red['R_age'],
                      y = red['R_Height_m'],
                      c = red['Winner'],
                      cmap = 'winter')
# Customize ax1
ax1.set(title = 'Effect of Height on Career Longevity',
        xlabel = 'Age',
        ylabel = 'Height')
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = 'Winner')
# Add a meanline
ax1.axhline(y = red['R_Height_m'].mean(),
            linestyle = '--');


# Add data to ax2
scatter = ax2.scatter(x = red['R_age'],
                      y = red['R_Reach_m'],
                      c = red['Winner'],
                      cmap = 'winter')
# Customize ax2
ax2.set(title = 'Effect of Reach on Career Longevity',
        xlabel = 'Age',
        ylabel = 'Reach')
# Add a legend to ax1
ax2.legend(*scatter.legend_elements(), title = 'Winner')
# Add a meanline
ax2.axhline(y = red['R_Reach_m'].mean(),
           linestyle = '--');


# Add data to ax3
scatter = ax3.scatter(x = red['R_Height_m'],
                      y = red['R_Reach_m'],
                      c = red['Winner'],
                      cmap = 'winter')
# Customize ax3
ax3.set(title = 'Importance of Range',
        xlabel = 'Height',
        ylabel = 'Reach')
# Add a legend to ax3
ax3.legend(*scatter.legend_elements(), title = 'Winner')
# Add a meanline
ax3.axhline(y = red['R_Reach_m'].mean(),
            linestyle = '--');


# Add a title to the figure
fig.suptitle("Winning Variables Analysis",
             fontsize = 16,
             fontweight = "bold");
```

# Machine Learning Model
1. Prepare data (rethink removing all missing data, in version 2, for model fitting more data)
2. Choose the right model (algorithm) for the problem
3. Fit the model and use it to make prediction on the data
4. Evaluating the model
5. Improve the model
6. Save and load the trained model

```python
# 1. Prepare data
# Create X (features: independent variables)
X = red.drop(['Winner', 'R_fighter'], axis = 1)

# Create y (labels: dependant variable)
y = red['Winner']
```

```python
# 2. Choose the right model and hyperparameter
# hyperparameters: dials on the model to tune its accuracy
from sklearn.ensemble import RandomForestClassifier

# default hyperparameter: n_estimator = 100
clf = RandomForestClassifier(n_estimators = 100) # Classification ML model
```

```python
# 3. Fit the model and use it to make prediction on the data
from sklearn.model_selection import train_test_split

# 80% of the data used for training the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Tells random forest classification model to find the patterns in the data
clf.fit(X_train, y_train)
```

```python
# Make a prediction
y_preds = clf.predict(X_test)
y_preds
```

```python
# 4. Evaluate the model on the training data and test data
print(clf.score(X_train, y_train), ', ', clf.score(X_test, y_test))
```

```python
# 3 different evaluators to test accuracy of predictions
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, y_preds)) # true labels vs predictions
```

```python
print(confusion_matrix(y_test, y_preds))
```

```python
print(accuracy_score(y_test, y_preds))
```

```python
# 5. Improve the model
# Trying different amounts of n_estimators ( a parameter to tune the model)
import numpy as np

np.random.seed(42)
for i in range(10, 100, 10):
    print(f'Trying model with {i} estimators...')
    clf = RandomForestClassifier(n_estimators = i).fit(X_train, y_train)
    print(f'Model accuracy on test set: {clf.score(X_test, y_test) * 100:2f}%')
    print('')
```

```python
# 6. Save and load the trained model
import pickle

pickle.dump(clf, open('ufc-random_forest_model', 'wb'))
```

```python
# import the saved model
loaded_model = pickle.load(open('ufc-random_forest_model', 'rb'))
loaded_model.score(X_test, y_test)
```
